# FLIR JSON to PVOC XML Annotations

```python
import os
import glob
from pathlib import Path
import json
from tqdm import tqdm
```

```python
files_path = Path('/home/karan/Downloads/FLIR_ADAS/video/Annotations')
files = files_path.glob('*.json')
files = list(files)
annotation_array = []
```

```python
id_to_cat = {'1': 'person', '2': 'bicycle', '3': 'car', '4': 'motorcycle', '5': 'airplane', '6': 'bus', '7': 'train', '8': 'truck', '9': 'boat', '11': 'fire hydrant', '10': 'traffic light','15': 'bench','1': 'traffic light', '17': 'cat', '18': 'dog'}
```

```python
files_names = []
for each in range(len(files)):
    with open(f'{files[each]}') as f:
        ann = json.load(f)
    filename = ann["image"]["file_name"]
    filename.strip('.')[0]
    files_names.append(filename)
```

```python
with open('/home/karan/Downloads/FLIR_ADAS/video/flir_vid.txt', 'w') as f:
    for each in files_names:
        print(each)
        f.write("%s\n" % each)
```

```python
def getXMLAnn(ann):
    st=""
    for i,b in enumerate(ann):
        if i == 0:
            st += f'\
        <object>\n \
                <name>{id_to_cat[b["category_id"]]}</name>\n \
                <pose>Frontal</pose>\n \
                <truncated>0</truncated>\n \
                <difficult>0</difficult>\n \
                <occluded>0</occluded>\n \
                <bndbox>\n \
                    <xmin>{b["bbox"][0]}</xmin>\n \
                    <xmax>{int(b["bbox"][0])+int(b["bbox"][2])}</xmax>\n \
                    <ymin>{b["bbox"][1]}</ymin>\n \
                    <ymax>{int(b["bbox"][1])+int(b["bbox"][3])}</ymax>\n \
                </bndbox>\n \
            </object> \n'
        else:
            st += f'\
            <object>\n \
                <name>{id_to_cat[b["category_id"]]}</name>\n \
                <pose>Frontal</pose>\n \
                <truncated>0</truncated>\n \
                <difficult>0</difficult>\n \
                <occluded>0</occluded>\n \
                <bndbox>\n \
                    <xmin>{b["bbox"][0]}</xmin>\n \
                    <xmax>{int(b["bbox"][0])+int(b["bbox"][2])}</xmax>\n \
                    <ymin>{b["bbox"][1]}</ymin>\n \
                    <ymax>{int(b["bbox"][1])+int(b["bbox"][3])}</ymax>\n \
                </bndbox>\n \
            </object> \n'
    return st

def toXML(ann):

    x = f'\
<annotation>\n \
    <folder>xmls</folder>\n \
    <filename>{ann["image"]["file_name"]}.jpg</filename>\n \
    <source>\n \
        <database>FLIR DATABASE</database>\n \
        <annotation>FLIR DATABASE</annotation> \n \
    </source>\n \
    <size>\n \
        <width>{ann["image"]["width"]}</width>\n \
        <height>{ann["image"]["height"]}</height>\n \
        <depth>3</depth>\n \
    </size>\n \
    <segmented>0</segmented>\n \
    {getXMLAnn(ann["annotation"])} \
</annotation>\n'
    return x
```

```python
from tqdm import tqdm
for each in tqdm(range(len(files))):
    with open(f'{files[each]}') as f:
        ann = json.load(f)
    with open(f'/home/karan/data/VOCdevkit2007/VOC2007/Annotations/{ann["image"]["file_name"]}.xml', 'w') as t:
        t.write(toXML(ann))
```
